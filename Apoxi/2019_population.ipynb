```python
def read_excel(filename):
    try:
        df = pd.read_excel(filename, header=1)
    except Exception as e:
        print(f"cannot read {filename}. Error: {e}")
    return df
```

```python
def find_header_index(dataframe):
    value = 'ΣΥΝΟΛΟ TOTAL'
    indexes = []
    columns_to_check = dataframe.columns.tolist()
    for item in columns_to_check:
        idx = dataframe[dataframe[item].str.contains(value, na=False)].index #.tolist()
        indexes.append(idx)
        return indexes[0]
```
